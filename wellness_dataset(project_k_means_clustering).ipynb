```python
x = df.drop(columns = ["stress_level_0_10"])
y = df["stress_level_0_10"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train , x_test , y_train , y_test = train_test_split(x , y , test_size = 0.2 , random_state = 42)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rf = RandomForestRegressor()
```

```python
rf.fit(x_train , y_train)
```

```python
y_pred = rf.predict(x_test)
```

```python
from sklearn.metrics import r2_score
r2_score(y_test , y_pred)
```

```python
clustering_data = df[["screen_time_hours" ,"work_screen_hours" , "leisure_screen_hours" , "sleep_hours" , "sleep_quality_1_5" , "stress_level_0_10" ,
                      "productivity_0_100" , "exercise_minutes_per_week" , "social_hours_per_week" , "mental_wellness_index_0_100"]]
```

```python
x = clustering_data.iloc[:, :].values
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
a = []

for i in range(1,11):
  b = KMeans(n_clusters=i , init="k-means++" , random_state=42)
  b.fit(x)
  a.append(b.inertia_)

plt.plot(range(1,11) , a)
plt.title("the elbow method graph")
plt.xlabel("no of cluster")
plt.ylabel("wcss_list")
plt.show()
```

```python
clustering_data
```

```python
from sklearn.preprocessing import MinMaxScaler
for i in clustering_data.columns:
  MinMaxScaler(i)

from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=3)
clusters = kmeans.fit_predict(clustering_data)
df["Segments"] = clusters
print(df.head(10))
```

```python
df["Segments"] = df["Segments"].map({0:"healthy balance" , 1:"high risk"})
```

```python
b = KMeans(n_clusters = 4 , init="k-means++" , random_state=42)
y_predict = b.fit_predict(x)
```

```python
# visualization the clusters
plt.scatter(x[y_predict==0 , 0] , x[y_predict==0 , 1] , s=100 , c="blue" , label="cluster1")
plt.scatter(x[y_predict==1 , 0] , x[y_predict==1 , 1] , s=100 , c="green" , label="cluster2")
plt.scatter(x[y_predict==2 , 0] , x[y_predict==2 , 1] , s=100 , c="red" , label="cluster3")
plt.scatter(x[y_predict==3 , 0] , x[y_predict==3 , 1] , s=100 , c="cyan" , label="cluster4")

plt.scatter(b.cluster_centers_[:,0] , b.cluster_centers_[:,1] , s=300 , c="yellow" , label="centroid")
plt.title("")
plt.xlabel("annual income(k$)")
plt.ylabel("spending score(1-100)")
plt.legend()
plt.show()
```
