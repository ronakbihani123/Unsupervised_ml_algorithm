```python
# how does pca work?
# lets suppose we have a dataset with many features
# 1. standardize the data --> since features may have different scales we normalize them .
# 2. compute the covariance metrics --> this tells us how variable are related (do they know together or independently ).
# 3. find eigan values and eigan vectors ---> eigan vectors = direction of new features(principal component)
#                                             eigan values = amount of variance (importance ) carried by each component.
# 4. sort by variance --> keep the components with the highest variance means they carried most important information .
# 5. for a new dataset --->  project original data onto the selected principal components
```

# What is PCA (Principal Component Analysis)?

# PCA is a dimensionality reduction technique in Machine Learning and Statistics.
# It transforms high-dimensional data into a smaller set of new variables called principal components, while still keeping most of the important information (variance) from the original dataset.

# Principal components are new features created as linear combinations of the original features.

# They are chosen so that the first component captures the maximum variance, the second captures the next maximum variance (orthogonal to the first), and so on.

# Why do we use PCA?

# We use PCA for several reasons:

# Dimensionality Reduction

# If your dataset has too many features (hundreds or thousands), models can become slow, overfit, or hard to visualize.

# PCA reduces the number of features while keeping most of the information.

# Remove Multicollinearity

# If features are highly correlated, PCA can combine them into independent principal components.

# Visualization

# PCA helps in visualizing high-dimensional data in 2D or 3D.

# For example, plotting the first two principal components can show clusters or patterns in data.

# Noise Reduction

# Since PCA keeps only the components with high variance, it removes components that may represent noise.

# Speeds up Algorithms

# With fewer features, algorithms like regression, clustering, or classification run faster.

Problems PCA Can Solve

High-Dimensional Data (Curse of Dimensionality)

Too many features make models slow, prone to overfitting, and hard to interpret.

PCA reduces dimensions while keeping most variance.

Example: Genetic data with thousands of gene expression features.

Multicollinearity (Correlated Features)

If features are highly correlated, regression or classification models may perform poorly.

PCA transforms correlated features into independent components.

Example: Economic indicators like GDP, income, and expenditure often overlap.

Visualization of High-Dimensional Data

Humans can only visualize data in 2D or 3D.

PCA projects high-dimensional data into 2D/3D for scatterplots.

Example: Reducing 784-pixel MNIST digit images into 2D for visualization.

Noise Reduction / Data Compression

Low-variance components often represent noise. Dropping them improves signal-to-noise ratio.

Example: Image compression (fewer features, but still recognizable images).

Speeding up Machine Learning Algorithms

Fewer features = less computation = faster training.

Example: Before clustering (like K-means), applying PCA speeds up convergence.

# Eigenvectors in PCA

# Eigenvectors = directions of maximum variance.

# They tell us the orientation of the new feature space (principal components).

# Each eigenvector is a line/axis along which the data varies the most.


# Eigenvalues in PCA

# Eigenvalues = magnitude of variance along eigenvectors.

# They tell us how important each eigenvector is.

# Larger eigenvalue → more variance captured → more information in that direction.

```python
import pandas as pd
import numpy as np
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv("/content/Social_Network_Ads (1).csv")
df.head(2)
```

```python
df = df.drop (columns=["User ID" , "Gender"
])
```

```python
df.head()
```

```python
x = df.drop(columns= ["Purchased"] , axis = 1)
y = df["Purchased"]
```

```python
x_train , x_test , y_train , y_test  = train_test_split(x,y,test_size=0.2,random_state=42)
pipe = Pipeline([
    ("scaler" , StandardScaler()) ,
    ("pca" , PCA(n_components=2)) ,
    ("classifier" , RandomForestClassifier(n_estimators=100 , random_state=42))
])
```

```python
pipe
```

```python
pipe.fit(x_train , y_train)
```

```python
y_pred = pipe.predict(x_test)
```

```python
acc = accuracy_score(y_test , y_pred)
print(acc)
```

```python
df = pd.read_csv("/content/userbehaviour.csv")
df.head()
```

```python
df.isnull().sum()
```

```python
from sklearn.preprocessing import OrdinalEncoder
oe = OrdinalEncoder()
df["Status"] = oe.fit_transform(df[["Status"]])
```

```python
df.head()
```

```python
df = df.drop(columns= ["userid"])
```

```python
x = df.drop(columns = ['Status'])
y = df["Status"]
```

```python
x_train , x_test , y_train , y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
new_pipe = Pipeline([
    ("scalar" , StandardScaler()),
    ("pca" , PCA(n_components=2)),
    ("classifier" , RandomForestClassifier(n_estimators=100 , random_state=42))
])
```

```python
new_pipe
```

```python
new_pipe.fit(x_train , y_train)
```

```python
y_pred = new_pipe.predict(x_test)
```

```python
accuracy_score(y_test , y_pred)
```
