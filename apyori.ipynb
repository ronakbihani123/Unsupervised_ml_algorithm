```python
# appriory algorithm in ml
# appriory is an association rule learning algorithm used in market basket analysis.
# its main goal is to find frequent item sets and then derive rules like if a customer buys bread they are likely to buy butter.
```

```python
# how does appriory work?
# ---> 1. find frequent item sets ---> calculate the suppot of item combinations like single item , pair or tripples. now keep only thosee a minimum threshold
# 2. generate association rules ---> from frequent item sets create rules like x --->y then evaluate them using confidence and lift
# key metrics (a). support ---> it measures how often an itemset appears in the dataset . support(x) = transactions containing x / total transactions
# ex -- > if 3 out of 10 customers buy milk suppot(milk) = 3/10
# (b). confidence---> it measures the probability of buying y given x , formula --> confidence(x) interval y = support (x)union y / suppot(x)
# ex --> out of 5 people who bought milk for also bought bread  confidencce(milk) to bread = 4/5
# (c) lift ---> it measure the strenght of a rule compared to random chance . formula --> lift(x) to y = confidence (x) to y / suppot(y)
# if lift is greater than 1 then x and y are positively related
# lift= 1 x and y are independent
# lift<1 x and y are negatively related
# ex --> if confidence (milk) to bread 0.8 or support .5 lift(x)to y = 0.8/0.5 = 1.6 --> meaning customer who buy milk are 1.6 times more likely to buy bread compared to random .
```

```python
# data processing
transactions = []
for i in range(0, 7501):
  transactions.append([str(df.values[i,j]) for j in range(0, 20)])

# training the dataset
from apyori import apriori
rules = apriori(transactions,        # (3 items * 7 days in week ) / total rows in dataset
                min_support = 0.003,
                min_confidence = 0.2,
                min_lift = 3,
                min_length = 2)
# visualizing the results
a = list(rules)
result = [list(a[i][0]) for i in range(0 , len(a))]
```
